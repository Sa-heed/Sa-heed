```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv")
df.head()
```

```python
# Checking the datatype
df.info()
```

```python
pd.set_option("display.max_rows", None)
```

```python
df = df.drop(columns=["date","lights"])
# df
```

```python
# Data preprosessing for regression
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
LM = LinearRegression()
scaler   =   MinMaxScaler()  
```

```python
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
normalised_df.head()
```

```python
# feature_df = normalised_df.drop(columns="Appliances")
# feature_df.head()
```

```python
# Splitting the data into training and testing for cross validation
from sklearn.model_selection import train_test_split as tts
x_train, x_test,y_train, y_test = tts(normalised_df["T2"].values.reshape(-1,1),normalised_df["T6"], train_size= 0.7 , random_state=42)
```

```python
model = LM.fit(x_train, y_train)
model
```

```python
predict = model.predict(x_test)
predict
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(predict,  y_test)
round(r_score, 2)                     # the r2 = 0.44 Question 12
```

```python
X = normalised_df.drop(columns=['Appliances'])
Y = normalised_df['Appliances']
```

```python
x_train, x_test, y_train, y_test = tts(X, Y, test_size=0.3, random_state=42)
LM1 = LinearRegression()
LM1.fit(x_train, y_train)
```

```python
predictions = LM1.predict(x_test)
predictions
```

```python
from sklearn.metrics import mean_absolute_error 
MAE_mr = mean_absolute_error(y_test, predictions)
round(MAE_mr, 2)
```

```python
# Question 14
SS = y_test - predictions # sum square error
RSS = np.sum(np.square(SS)) # Residual sum of square
round(RSS,2)
```

```python
#question 15

from sklearn.metrics import mean_squared_error 
RMSE = np.sqrt(mean_squared_error(y_test, predictions))
round(RMSE, 3)
```

```python
#question 16
from sklearn import metrics
coefficent = metrics.r2_score(y_test, predictions)
round(coefficent,2)
```
